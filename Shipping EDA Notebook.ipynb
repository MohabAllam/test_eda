```python
from sqlalchemy import create_engine

connection = create_engine("mysql+pymysql://root:2416734@localhost:3306/ecommdb")

customers = pd.read_sql(sql= 'SELECT * FROM customers;',con= connection)
orders = pd.read_sql(sql= 'SELECT * FROM orders;',con= connection)
products = pd.read_sql(sql= 'SELECT * FROM products;',con= connection)
sales = pd.read_sql(sql= 'SELECT * FROM sales;',con= connection)
```

```python
orders
```

```python
df = pd.merge(left= customers, right= orders, how='inner', on= 'customer_id')
df.head()
```

```python
df = pd.merge(left= df, right= sales, how='inner', on= 'order_id')
df.head()
```

```python
df = pd.merge(left= df, right= products, how='inner', left_on= 'product_id', right_on = 'product_ID')
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns= {'quantity_x' : 'quantity', 'quantity_y' : 'Stock'})
df.head()
```

## Data Cleaning

```python
# Drop Unnecessary Columns

df.drop(['customer_id', 'home_address', 'zip_code', 'country', 'order_id', 'product_id', 'product_ID', 'description'], axis= 1, inplace= True)
```

```python
# Check duplicates
df.duplicated().sum()
```

```python
df.head(1)
```

```python
# Convert to datetime

df['order_date'] = pd.to_datetime(df['order_date'], dayfirst= True)
df['delivery_date'] = pd.to_datetime(df['delivery_date'], dayfirst= True)
```

```python
df.info()
```

- In depth check for numerical columns

```python
num_cols = df.select_dtypes(include= 'number').columns.drop('sales_id')
num_cols
```

```python
for col in num_cols:

    px.histogram(data_frame= df, x= col, title= col).show()
```

- In depth check for categorical columns

```python
cat_cols = df.select_dtypes(include= 'object').columns.drop('customer_name')
cat_cols
```

```python
for col in cat_cols:

    print(col)
    print(df[col].nunique())
    print(df[col].unique())
    print('-' * 100)
```
